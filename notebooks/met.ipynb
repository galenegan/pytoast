```python
import glob
from pytoast.atmosphere.met import Met
```

```python
# Files from test directory
files = glob.glob("../py-tests/src/atmosphere/testdata/*.csv")
files.sort()

# pytoast_name: source_name pairs (variable names as stored in the .csv file)
# Here we specify a list of names corresponding to the two different measurement heights
# z in this case is passed in as a variable rather than an initializer to the class
name_map = {
    "time": "time",
    "p": ["atmospheric_pressure_0", "atmospheric_pressure_1"],
    "t": ["air_temperature_0", "air_temperature_1"],
    "rh": ["relative_humidity_0", "relative_humidity_1"],
    "z": ["z_met_0", "z_met_1"]
}

# Initializing Met
met = Met(files=files, name_map=name_map, fs=1.0 / (20 * 60))
```

```python
# Loading the burst and calculating derived quantities
burst = met.load_burst(0)
data = met.derive(burst)
```
