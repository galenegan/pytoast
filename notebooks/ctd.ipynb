```python
import glob
import numpy as np
from pytoast.ocean.ctd import CTD
```

```python
# Files from test directory
files = glob.glob("../py-tests/src/ocean/ctd/testdata/*.mat")
files.sort()

# pytoast_name: source_name pairs (variable names as stored in the .mat file)
name_map = {"sp": "sal", "t": "temp", "p": "depth", "time": "time"}

# Initializing CTD object. Two instrument depths -> derive() will also return N2
ctd = CTD(files=files, name_map=name_map, fs=1.0, z=[2.5, 3.3], z_convention="depth")
```

```python
# Loading a burst (sp, t, p, time) and deriving thermodynamic quantities in place
burst = ctd.load_burst(0)
data = ctd.derive(burst)

print("Keys added by derive:")
print(sorted(data.keys()))
```

```python
# Burst-mean values of a few derived quantities
for key in ["sa", "ct", "rho", "sigma0", "sound_speed", "cp", "nu"]:
    print(f"{key:12s} mean = {np.mean(data[key]):.4e}")

# N2 (buoyancy frequency squared) is computed between adjacent depths, so it has
# one fewer row than the number of instrument heights
print(f"\nN2 shape (n_heights-1, n_samples): {data['N2'].shape}")
print(f"N2 mean (1/s^2): {np.mean(data['N2']):.3e}")
```
